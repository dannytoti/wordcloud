# Word clouds Visualiton using the word_cloud package

Word cloud can be useful for high level analyse of textual dataset to get most used words in the dataset

```python
pip install wordcloud
```

```python
# now that we have install the package let's import the modules we need
from wordcloud import STOPWORDS, WordCloud
```

```python
#next let's get our data (the data set was provided by IBM)
import urllib

# open the file and read it into a variable alice_novel
alice_novel = urllib.request.urlopen('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/alice_novel.txt').read().decode("utf-8")
```

```python
stopword = set(STOPWORDS)
```

```python
#initiate the word cloud object

alice_wc = WordCloud(background_color='white',
                    max_words=2000,
                    stopwords=stopword)

alice_wc.generate(alice_novel)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# We can see that some of the most used from the dataset are ALICE, SAID, LITTLE, ONE, etc...
```

```python
#now let's make of word cloud a little bigger
```

```python
fig=plt.figure(figsize=(15,8))
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Another cool thing is that we can used a mask with the word cloud such that the words take the shape of the image. Let's do that next

```python
import numpy as np
from PIL import Image
alice_mask = np.array(Image.open(urllib.request.urlopen('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/labs/Module%204/images/alice_mask.png')))
```

```python
# let's visualize the image
fig = plt.figure(figsize=(15,10))

plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()

# Nice !
```

```python
# now let's combine our word cloud to the mask image
alice_wc = WordCloud(stopwords=stopword,mask=alice_mask, background_color='white', max_words=2000)

alice_wc.generate(alice_novel)

fig=plt.figure(figsize=(14,18))

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
